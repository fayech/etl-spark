```python
import time
import datetime
from pyspark.sql.functions import lit,unix_timestamp
```

```python
datetime
```

```python
timestamp = datetime.datetime.fromtimestamp(time.time()).strftime('%Y-%m-%d %H:%M:%S')
```

```python
newdf = df.withColumn('DATDEBVLD',unix_timestamp(lit(timestamp),'yyyy-MM-dd HH:mm:ss').cast("timestamp")).withColumn('DATCHG',unix_timestamp(lit(timestamp),'yyyy-MM-dd HH:mm:ss').cast("timestamp"))
                                                                                                                     
```

```python
findata=newdf.withColumn("DATFINVLD",lit(999))
```

```python
findata.show()
```

```python
findata.printSchema()
```

```python
findata.coalesce(1).write.csv("use.csv")
```
